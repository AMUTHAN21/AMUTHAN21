```python
start_date="01-01-2001"
end_date="01-01-2005"
timelim=(data.index>=start_date)&(data.index<=end_date)
datanew=data.loc[timelim]
```

```python
datanew.plot(figsize=(12,6))
```

```python
dataquarterly=data.rolling(window=3).mean()
dataquarterly.plot(figsize=(14,6))
```

```python
datayearly=data.rolling(window=12).mean()
datayearly.plot(figsize=(14,6))
```

```python
import pandas as pd
```

```python
data=pd.read_csv('/content/Course+DataSet.csv',index_col="DATE",parse_dates=True)
```

```python
import matplotlib.pyplot as plt
```

```python
start_date="01-01-2010"
end_date="01-01-2017"
timelin=( data.index >= start_date) & ( data.index < end_date)
datatimelin=data.loc[timelin]
datatimelin.plot(figsize=(12,6))
xcoords=['01-01-2010','01-01-2011','01-01-2012','01-01-2013','01-01-2014','01-01-2015','01-01-2015']
for xc in xcoords:
  plt.axvline(x=xc,color='black',linestyle='--')
```
